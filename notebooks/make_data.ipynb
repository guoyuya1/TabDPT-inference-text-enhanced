```python
import pandas as pd
import numpy as np
import seaborn as sns
from pathlib import Path
```

```python
df = pd.read_csv("../data/processed_data/bitcoin_final_with_embeddings_lag_7.csv")
df.head()
```

```python
df.columns
```

```python
dataset_lag = 7
emb_df = df[[f"embedding_summary_gpt-5-mini_lag{i}" for i in range(1, dataset_lag + 1)]]
emb_matrix = emb_df.applymap(lambda x: np.array(x[1:-1].split(",")).astype(float))
```

```python
def get_scores(all_vectors, dir1):
    # all_vectors: N x D matrix of N vectors of dimension 4096
    # dir1: D-dimensional vector
    dots = all_vectors @ dir1
    dots_matrix = np.stack([dots, -dots], axis=1)
    exps = np.exp(dots_matrix)
    softmax_scores = exps / np.sum(exps, axis=1, keepdims=True)
    return softmax_scores

def create_lag_matrix(v, d):
    """
    v: Input vector of shape (N,) or (N, 1)
    d: Number of lags to create
    Returns: Matrix of shape (N, d)
    """
    v = v.ravel()
    N = v.shape[0]
    
    # Initialize matrix with NaNs
    lag_matrix = np.full((N, d), np.nan)
    
    for i in range(d):
        # The i-th lag shifts the data down by (i+1)
        lag_val = i + 1
        if lag_val < N:
            # Fill the column from index 'lag_val' to the end
            lag_matrix[lag_val:, i] = v[:-lag_val]
            
    return lag_matrix

null_vector = emb_matrix["embedding_summary_gpt-5-mini_lag1"][0]
rng = np.random.default_rng(seed=42)
r = rng.standard_normal(len(null_vector))
dir1 = r - (np.dot(r, null_vector) / np.dot(null_vector, null_vector)) * null_vector
```

```python
scale = 1
all_scores = get_scores(np.stack(emb_matrix["embedding_summary_gpt-5-mini_lag1"].values), dir1)
direction = scale * (all_scores[:, 0] - all_scores[:, 1])
sns.displot(direction,kde=True)

# #Generate lagged features
direction_lags = create_lag_matrix(direction, dataset_lag)
direction_lags[np.where(np.isnan(direction_lags))] = 0.0
direction_lags_df = pd.DataFrame(direction_lags, columns=[f"direction_lag{i}" for i in range(1, dataset_lag + 1)])
df = pd.concat([df, direction_lags_df], axis=1)
```

```python
df_interaction = df.copy()
df_additive  = df.copy()

target_lag = 3
input_cols = ["Open", "High", "Low", "Close", "Adj_Close"]
init_dict = {"target_lag1": 0.2, "target_lag2": 0.1, "target_lag3": 0.1}
params_scale = 0.5
```

## Target generation with Interaction Terms

```python
#simple interaction terms (Can be more complicated in future)
trainable_cols_interaction = []
for col in input_cols:
    for lag1 in range(1, dataset_lag):
        for lag2 in range(lag1 + 1, dataset_lag + 1):
            df_interaction[f"{col}_direction_lag{lag1}_{lag2}"] = df_interaction[f"{col}_lag{lag1}"] * df_interaction[f"direction_lag{lag2}"]
            if (lag1<=target_lag) and (lag2<=target_lag): #Only include lag terms until target_lag for interaction terms to keep it simple and avoid too many features
                trainable_cols_interaction.append(f"{col}_direction_lag{lag1}_{lag2}")
        if lag1 <= target_lag:
            trainable_cols_interaction.append(f"{col}_lag{lag1}")
    if dataset_lag <= target_lag:
        trainable_cols_interaction.append(f"{col}_lag{dataset_lag}")

#generate synthetic target variable
text_scale = 0.2
rng = np.random.default_rng(seed=42)
noise = rng.standard_normal(len(df))
random_params = np.random.uniform(0, 1, size=len(trainable_cols_interaction))
param_dict_interact = init_dict.copy()
random_params = random_params / np.linalg.norm(random_params) * params_scale
for i,col in enumerate(trainable_cols_interaction):
    if "direction_lag" in col:
        param_dict_interact[col] = random_params[i]*text_scale
    else:
        param_dict_interact[col] = random_params[i]
```

## Target generation with additive terms only

```python
# Create signal from text features
n_points = len(df)
target_mean = 5000
standard_deviation = 0  # Controls how "spread out" the noise is

text_signal = direction * np.random.normal(loc=target_mean, scale=standard_deviation, size=n_points) + 500

#Generate target variable for additive model
text_signal_lags = create_lag_matrix(text_signal, dataset_lag)
text_signal_lags_df = pd.DataFrame(text_signal_lags, columns=[f"text_signal_lag{i}" for i in range(1, dataset_lag + 1)])
df_additive = pd.concat([df_additive, text_signal_lags_df], axis=1)
```

```python
trainable_cols_additive = [col for col in trainable_cols_interaction if "direction" not in col]
trainable_cols_additive+= [f"text_signal_lag{j}" for j in range(1, target_lag + 1)]

#generate synthetic target variable
noise = rng.standard_normal(len(df))
random_params = np.random.uniform(0, 1, size=target_lag)
random_params = random_params / np.linalg.norm(random_params) * 0.85
param_dict_additive = {f"text_signal_lag{i}": random_params[i-1] for i in range(1, target_lag + 1)}
for i,col in enumerate(param_dict_interact.keys()):
    if "direction_lag" not in col:
        param_dict_additive[col] = param_dict_interact[col]
```

## Generate data

```python
start_val = 13000
end_val = 17000
df_interaction["target"] = np.nan
df_interaction["no_text_target"] = np.nan
df_additive["target"] = np.nan
target = list(np.random.uniform(start_val, end_val, size=target_lag)) # Start with random values for the first 'target_lag' entries
no_text_target = target.copy()
additive_text_target = target.copy()

df_interaction["target"].iloc[:target_lag] = target
df_interaction["no_text_target"].iloc[:target_lag] = no_text_target
df_additive["target"].iloc[:target_lag] = additive_text_target

for i in range(target_lag, len(df_interaction)):
    lagged_values_interactive = [target[i - j] for j in range(1, target_lag + 1)]
    lagged_values_no_text = [no_text_target[i - j] for j in range(1, target_lag + 1)]
    lagged_values_additive = [additive_text_target[i- j] for j in range(1, target_lag + 1)]
    
    feature_values_interactive = df_interaction.loc[i, trainable_cols_interaction].values
    feature_values_additive = df_additive.loc[i, trainable_cols_additive].values

    target_value = sum(param_dict_interact[f"target_lag{j}"] * lagged_values_interactive[j-1] for j in range(1, target_lag + 1))
    target_value += sum(param_dict_interact[trainable_cols_interaction[k]] * feature_values_interactive[k] for k in range(len(trainable_cols_interaction)))
    no_text_target_value = sum(param_dict_interact[f"target_lag{j}"] * lagged_values_no_text[j-1] for j in range(1, target_lag + 1))
    no_text_target_value += sum(param_dict_interact[trainable_cols_interaction[k]] * feature_values_interactive[k] for k in range(len(trainable_cols_interaction)) if "direction" not in trainable_cols_interaction[k])
    additive_text_target_value = sum(param_dict_additive[f"target_lag{j}"] * lagged_values_additive[j-1] for j in range(1, target_lag + 1))
    additive_text_target_value += sum(param_dict_additive[trainable_cols_additive[k]] * feature_values_additive[k] for k in range(len(trainable_cols_additive)))
    
    target_value += noise[i] * 0  # Add some noise, for now no noise to keep it deterministic
    no_text_target_value += noise[i] * 0
    additive_text_target_value += noise[i] * 0

    target.append(target_value)
    no_text_target.append(no_text_target_value)
    additive_text_target.append(additive_text_target_value)

df_interaction["target"] = target
df_interaction["no_text_target"] = no_text_target
df_additive["target"] = additive_text_target

#Plot the target variable and close price to visualize the synthetic data
from matplotlib import pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(np.arange(len(df_interaction)), df_interaction["target"], label="Target")
plt.plot(np.arange(len(df_interaction)), df_interaction["no_text_target"], label="No Text Target")
plt.plot(np.arange(len(df_additive)), df_additive["target"], label="Additive Text Target")
plt.plot(np.arange(len(df_interaction)), df_interaction["Adj_Close"], label="Adj Close Price")
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.title("Synthetic Target Variable Over Time")
plt.legend()
plt.show()
```

```python
#Save the synthetic dataset with lagged features and target variable
target_lagged = create_lag_matrix(df_interaction["target"].values, dataset_lag)
target_lags_df = pd.DataFrame(target_lagged, columns=[f"target_lag{i}" for i in range(1, dataset_lag + 1)])
synthetic_df = pd.concat([df_interaction, target_lags_df], axis=1)
trainable_cols_interaction.extend([f"target_lag{i}" for i in range(1, target_lag + 1)])
synthetic_df.to_csv("../data/processed_data/bitcoin_synthetic_data_interactive.csv", index=False)

target_lagged_additive = create_lag_matrix(df_additive["target"].values, dataset_lag)
target_lags_df = pd.DataFrame(target_lagged_additive, columns=[f"target_lag{i}" for i in range(1, dataset_lag + 1)])
synthetic_df_additive = pd.concat([df_additive, target_lags_df], axis=1)
trainable_cols_additive.extend([f"target_lag{i}" for i in range(1, target_lag + 1)])
synthetic_df_additive.to_csv("../data/processed_data/bitcoin_synthetic_data_additive.csv", index=False)
```

### Baseline

```python
# Train-test split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

clean_df = synthetic_df.iloc[dataset_lag:]
clean_df_additive = synthetic_df_additive.iloc[dataset_lag:]
final_trainable_cols = trainable_cols_interaction.copy()
final_trainable_cols_additive = trainable_cols_additive.copy()


train_size = 200
train_df = clean_df.iloc[:train_size]
test_df = clean_df.iloc[train_size:]

train_df_additive = clean_df_additive.iloc[:train_size]
test_df_additive = clean_df_additive.iloc[train_size:]

### Baseline
print("\nLinear Model: Considering full feature set with interactions and lags")
linmod = LinearRegression()
linmod.fit(train_df[final_trainable_cols], train_df["target"])
y_train_pred = linmod.predict(train_df[final_trainable_cols])
y_test_pred = linmod.predict(test_df[final_trainable_cols])
print("Train MAE:", mean_absolute_error(train_df["target"], y_train_pred))
print("Train R2:", r2_score(train_df["target"], y_train_pred))
print("Test MAE:", mean_absolute_error(test_df["target"], y_test_pred))
print("Test R2:", r2_score(test_df["target"], y_test_pred))

print("\nLinear Model: No terms from text direction (to test if model is actually using text info)")
new_trainable_cols = [col for col in final_trainable_cols if 'direction' not in col]
linmod_notext = LinearRegression()
linmod_notext.fit(train_df[new_trainable_cols], train_df["target"])
y_train_pred = linmod_notext.predict(train_df[new_trainable_cols])
y_test_pred = linmod_notext.predict(test_df[new_trainable_cols])

print("Train MAE:", mean_absolute_error(train_df["target"], y_train_pred))
print("Train R2:", r2_score(train_df["target"], y_train_pred))
print("Test MAE:", mean_absolute_error(test_df["target"], y_test_pred))
print("Test R2:", r2_score(test_df["target"], y_test_pred))


print("\nLinear Model: Additive terms from text direction (to test if model is actually using text info)")
linmod_additive = LinearRegression()
linmod_additive.fit(train_df_additive[final_trainable_cols_additive], train_df_additive["target"])
y_train_pred = linmod_additive.predict(train_df_additive[final_trainable_cols_additive])
y_test_pred = linmod_additive.predict(test_df_additive[final_trainable_cols_additive])

print("Train MAE:", mean_absolute_error(train_df_additive["target"], y_train_pred))
print("Train R2:", r2_score(train_df_additive["target"], y_train_pred))
print("Test MAE:", mean_absolute_error(test_df_additive["target"], y_test_pred))
print("Test R2:", r2_score(test_df_additive["target"], y_test_pred))

print("\nLinear Model: Additive terms from text direction, no text signal used (to test if model is actually using text info)")
new_trainable_cols = [col for col in final_trainable_cols_additive if 'text' not in col]
linmod_additive = LinearRegression()
linmod_additive.fit(train_df_additive[new_trainable_cols], train_df_additive["target"])
y_train_pred = linmod_additive.predict(train_df_additive[new_trainable_cols])
y_test_pred = linmod_additive.predict(test_df_additive[new_trainable_cols])
print("Train MAE:", mean_absolute_error(train_df_additive["target"], y_train_pred))
print("Train R2:", r2_score(train_df_additive["target"], y_train_pred))
print("Test MAE:", mean_absolute_error(test_df_additive["target"], y_test_pred))
print("Test R2:", r2_score(test_df_additive["target"], y_test_pred))
```
