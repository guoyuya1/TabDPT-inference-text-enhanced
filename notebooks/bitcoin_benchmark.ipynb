### Load dataset

```python
import pandas as pd
df_path = './data/source_data/bitcoin/bitcoin_price_news.csv'
label = ['Close']
features = ['Open',	'High',	'Low',	'Volume']
train_split=0.6
valid_split=0.2
input_window_size = 10
output_window_size = 1
df = pd.read_csv(df_path)[label+features][:1000]
n = df.shape[0]
train_n = int(n * train_split)
valid_n = int(n * (train_split + valid_split))

train_df = df.iloc[:train_n]
valid_df = df.iloc[train_n:valid_n]
test_df = df.iloc[valid_n:]

test_true = test_df[label[0]].iloc[input_window_size:].to_list()
```

#### Nlinear

```python
# nlinear
import numpy as np
from darts.models import NLinearModel
from darts import TimeSeries
from sklearn.metrics import mean_absolute_error, root_mean_squared_error

train_series = TimeSeries.from_values(train_df[features + label].values)

model_NLinearModel = NLinearModel(input_chunk_length=input_window_size,
                                  output_chunk_length=output_window_size, 
                                  pl_trainer_kwargs={"accelerator": "gpu", "devices": 1}, )
model_NLinearModel.fit(train_series)

# predict
pred = []
for i in range(len(test_df)-input_window_size):
    test_series = TimeSeries.from_values(np.array(test_df[features + label].iloc[i:i+input_window_size]))
    predictions = model_NLinearModel.predict(n=output_window_size, series=test_series, verbose=0).values().tolist()
    pred.append(predictions)

# get prediction
temp_pred = []
for i in pred:
    # collect temp
    temp_pred.append(i[0][0])

# calculate metrics
print('MAE', mean_absolute_error(temp_pred, test_true))
print('RMSE', root_mean_squared_error(temp_pred, test_true))
```

#### Dlinear

```python
# dlinear
import numpy as np
from darts.models import DLinearModel
from darts import TimeSeries
from sklearn.metrics import mean_absolute_error, root_mean_squared_error

train_series = TimeSeries.from_values(train_df[features + label].values)

model_DLinearModel = DLinearModel(input_chunk_length=input_window_size,
                                  output_chunk_length=output_window_size, 
                                  pl_trainer_kwargs={"accelerator": "gpu", "devices": 1}, )
model_DLinearModel.fit(train_series)

# predict
pred = []
for i in range(len(test_df)-input_window_size):
    test_series = TimeSeries.from_values(np.array(test_df[features + label].iloc[i:i+input_window_size]))
    predictions = model_DLinearModel.predict(n=output_window_size, series=test_series, verbose=0).values().tolist()
    pred.append(predictions)

# get prediction
temp_pred = []
for i in pred:
    # collect temp
    temp_pred.append(i[0][0])

# calculate metrics
print('MAE', mean_absolute_error(temp_pred, test_true))
print('RMSE', root_mean_squared_error(temp_pred, test_true))
```
